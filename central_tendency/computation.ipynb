```python
import numpy as np
import matplotlib.pyplot as plt

datasheet = np.array([1,2,2,3,4,5,6])

mean = np.mean(datasheet)

sample_variance = np.var(datasheet, ddof=0)
sample_std = np.std(datasheet, ddof=1)

plt.figure(figsize=(8,5))
plt.scatter(range(1, len(datasheet)+1), datasheet, label = "Testing Datasheet", color = "blue")
plt.axhline(mean, color = "red", linestyle = "--", label = f"Mean(${mean:.2f})")

for index, val in enumerate(datasheet, start = 1):
    plt.plot([index, index], [mean, val], color = "gray", linestyle = ":")


plt.title('Sample Datasheet')
plt.xlabel("Days")
plt.ylabel("Expenses")
plt.legend()
plt.grid(True)
plt.show(), sample_variance, sample_std
```
